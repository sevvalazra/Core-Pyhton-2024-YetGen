# Problem 1
Vehicle sınıfından miras alan bir Bus sınıfı oluşturun. Bus.seating_capacity()'nin kapasite bağımsız değişkenine varsayılan olarak 50 değerini verin.

```
Output:
> The seating capacity of a bus is 50 passengers
```


````python
class Vehicle():
    def __init__(self,name,speed,milage):
        self.name=name
        self.speed=speed
        self.milage=milage
    def seating_capacity(self):
         return f"The seating capacity of a bus is {self._seating_capacity} passengers"
class Bus(Vehicle):
    def __init__(self, name, speed, milage, seating_capacity=50):
        super().__init__(name, speed, milage)
        self._seating_capacity = seating_capacity
    

bus1 = Bus("School Volvo",180, 17)
print(bus1.seating_capacity())      
````
