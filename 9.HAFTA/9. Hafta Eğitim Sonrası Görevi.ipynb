# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python

import random
import time


class Kumanda():
    def _init_(self,tv_durum="Kapalı",tv_ses=0,kanal_listesi=["Trt"],kanal="Trt"): 
        self.tv_durum=tv_durum
        self.tv_ses=tv_ses
        self.kanal_listesi=kanal_listesi
        self.kanal=kanal
    
    def tv_ac(self):
       
        if (self.tv_durum=="Açık"):
            print("Televizyon zaten açık.")
        else:
            print("Televizyon Açılıyor...")
            self.tv_durum="Açık"

    def tv_kapat(self):
        if (self.tv_durum=="Kapalı"):
            print("Televizyon zaten kapalı.")
        else:
            print("Televizyon Kapanıyor...")
            self.tv_durum="Kapalı"
    def ses_ayarlari(self):
        # YetNot: while döngüsüne alalım.
        while True:
            # YetNot: cevap islimli bir değişken oluşturup kullanıcıdan ses azaltmak için < yazmasını, ses artırmak için >, çıkmak için çıkış yazmasını isteyelim.
            cevap=input("ses azaltmak için < yaz, ses artırmak için > yaz, çıkmak için çıkış yaz")
            # YetNot: cevap < ise tv_ses 0 değilse sesi bir azaltıp ekrana tv_ses değerini bastıralım.
            if cevap == "<":
                if (self.tv_ses != 0):
                    self.tv_ses -=1 
                    print("Ses:",self.tv_ses)
           # YetNot: cevap > ise tv_ses 20 değilse sesi bir artırıp ekrana tv_ses değerini bastıralım.
            elif (cevap==">"):
                if (self.tv_ses!=20):
                    self.tv_ses+=1
                    print("Ses:",self.tv_ses)
           # YetNot: Diğer durumlarda ses güncellendi yazıp değeri ekrana yazdıralım.
            else:
                print("Ses güncellendi. Ses:",self.tv_ses)
                break
    # YetNot: kanal_ekle isimli bir fonksiyon oluşturup self ve kanal_ismi parametresi gönderelim.
    def kanal_ekle(self,kanal_ismi):
        # YetNot: Ekrana kanal ekleniyor yazdırıp 1 saniye bekleyelim.
        print("Kanal Ekleniyor...")
        time.sleep(1)
        # YetNot: kanal_listesine kanalı ekleyip kanal eklendi yazdıralım.
        self.kanal_listesi.append(kanal_ismi)
        print("Kanal eklendi!")
    # YetNot: rastgele_kanal isimli fonksiyon oluşturp self parametresini gönderelim.
    def ratgele_kanal(self):
        # YetNot: rastgele isimli bir değişken oluşturup random modülü yardımıyla 0 ile kanal_listesi - 1 arasında bir sayı seçtirelim.
        rastgele=random.randint(0,len(self.kanal_listesi)-1)
        # YetNot: Seçilen indeksteki elemanı kanala atayalım.
        self.kanal=self.kanal_listesi[rastgele]
        # YetNot: Kanal ismini ekrana bastıralım.
        print("Şu anki kanal:",self.kanal)
    # Yetnot: _len_ isimli fonksiyon oluşturup kanal sayısını return edelim.
    def _len_(self):
        return len(self.kanal_listesi)
    # Yetnot: _str_ isimli fonksiyon oluşturup format ile kanal durumunu öğrenelim.
    def _str_(self):
        return f"Tv Durumu:{self.tv_durum}\n Tv Ses:{self.tv_ses}\n Kanal Listesi:{self.kanal_listesi}\n Şu an bulunulan kanal:{self.kanal}"
# YetNot: kumanda nesnesi üretelim.
kumanda=Kumanda()



print("""
Televizyon Uygulaması
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri
Çıkmak için 'q' ya basın.
""")

# YetNot: While döngüsü kuralım.
while True:
    # YetNot: islem adında bir değişken oluşturup input ile işlem alalım.
    islem=input("Lütfen yapmak istediğiniz işlemi giriniz.")
    # YetNot: islem q ise Program Sonlandırılıyor yazdıralım.
    if islem=="q":
        print("Program Sonlandırılıyor.")
        break
    # YetNot: islem 1 ise kumanda sınıfından tv_ac fonksiyonunu çağıralım.
    elif islem=="1":
        kumanda.tv_ac()
    # YetNot: islem 2 ise kumanda sınıfından tv_kapat fonksiyonunu çağıralım.
    elif islem =="2":
        kumanda.tv_kapat()
    # YetNot: islem 3 ise kumanda sınıfından ses_ayarlari fonksiyonunu çağıralım.
    elif islem == "3":
        kumanda.ses_ayarlari()
    # YetNot: islem 4 ise kanal isimlerini , ile ayırarak girmesini isteyelim.
    # , 'den sonra ayıralım ve for döngüsü yardımıyla kanal_ekle fonksiyonuna gönderelim.
    elif islem == "4":
        kanal_isimleri=input("Kanal isimlerini ',' ile ayırarak girin.")
        kanal_listesi=kanal_isimleri.split(",")
        # YetNot: Ekrana kanal ekleniyor yazdırıp 1 saniye bekleyelim.
        for i in kanal_listesi:
            kumanda.kanal_ekle(i)
        # YetNot: kanal_listesine kanalı ekleyip kanal eklendi yazdıralım.
    
    # YetNot: islem 5 ise kanal sayısını yazdıralım.
    elif islem=="5":
        print("Kanal sayısı:",len(kumanda))
    # YetNot: islem 6 ise kumanda sınıfından rastgele_kanal fonksiyonunu çağıralım.
    elif islem == "6":
        kumanda.ratgele_kanal()
    # YetNot: islem 7 ise kumanda sınıfını çağıralım.
    elif islem=="7":
        print(kumanda)
    # YetNot: Yanlış işlemse geçersiz işle
    else:
        print("Geçersiz işlem.")
```
