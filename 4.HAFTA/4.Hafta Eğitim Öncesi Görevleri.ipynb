# Problem 3

3 öğrenciden oluşan bir öğrenci not sözlüğü oluşturun. Bu sözlükte öğrencilerin notları value olarak bir listede toplansın.

Kullanıcıya hangi öğrencinin notlarını görmek istediğini sorun. Öğrencinin notu görüntülendiğinde program sonunda şöyle bir çıktı elde etmelisiniz:

```
Lütfen notlarını görmek istediğiniz öğrencinin adını girin: Mehmet

Mehmet isimli öğrencinin      1.Sınav Notu:72
                              2.Sınav Notu:66
                              3.Sınav Notu:48
Not Ortalaması: 62.0


````python
students = {
    "Azra": [30, 43, 85],
    "Mehmet": [72, 66, 48],
    "Enes": [23, 100, 59]
}

student = input("Öğrencinin adı: ")

if student in students:
    not1 = students[student][0]
    not2 = students[student][1]
    not3 = students[student][2]
    average = (not1 + not2 + not3) / 3

    
    print(
        f"""
        {student} isimli öğrencinin notları:
          1. sınav notu: {not1}
          2. sınav notu: {not2}
          3. sınav notu: {not3}
          Not Ortalaması: {average}  
        """
    )
else:
    print(f"{student} adlı öğrenci bulunamadı.")
````
